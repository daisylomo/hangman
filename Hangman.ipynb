```python
import time
from getpass import getpass

word = str(getpass("Secret word: ")).lower()
name = input("Name: ")
print("Hello " + name, ",Play now!")

guesses = ""
turns = 5

while turns > 0:
        failed = 0
    for char in word:
        if char in guesses:
            print(char, end = " ")
        else:
            print("_", end = " ")
            failed += 1
    if failed == 0:
        print("Winner!")
        break       
    guess = input("Guess: ").lower()
    guesses += guess
    print(guesses)
    if guess not in word:
        turns -= 1
        print("Wrong")
        print("You have",+turns, "turns")
        if turns == 0:
            print("You lose!")
```
